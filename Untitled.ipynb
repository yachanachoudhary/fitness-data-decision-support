```python
# Import pandas
import pandas as pd

# Define file containing dataset (use full correct path)
runkeeper_file = r"C:\Users\HP\OneDrive\Desktop\internship project\Analyze Your Runkeeper Fitness Data\datasets\cardioActivities.csv"

# Create DataFrame with parse_dates and index_col parameters
df_activities = pd.read_csv(runkeeper_file, parse_dates=['Date'], index_col='Date')

# First look at exported data: select sample of 3 random rows
display(df_activities.sample(3))

# Print DataFrame summary
print(df_activities.info())
```

```python
# Define list of columns to be deleted
cols_to_drop = ['Friend\'s Tagged','Route Name','GPX File','Activity Id','Calories Burned', 'Notes']

# Delete unnecessary columns
df_activities.drop(cols_to_drop, axis=1, inplace=True)

# Count types of training activities
display(df_activities['Type'].value_counts())

# Rename 'Other' type to 'Unicycling'
df_activities['Type'] = df_activities['Type'].replace('Other', 'Unicycling')

# Count missing values for each column
display(df_activities.isnull().sum())
```

```python
# Calculate sample means for heart rate for each training activity type 
mean_hr_running = df_activities[df_activities['Type'] == 'Running']['Average Heart Rate (bpm)'].mean()
mean_hr_cycling = df_activities[df_activities['Type'] == 'Cycling']['Average Heart Rate (bpm)'].mean()

# Split the main DataFrame into activity-specific DataFrames
running_data = df_activities[df_activities['Type'] == 'Running'].copy()
walking_data = df_activities[df_activities['Type'] == 'Walking'].copy()
cycling_data = df_activities[df_activities['Type'] == 'Cycling'].copy()

# Fill missing heart rate values using calculated means (safe version)
walking_data['Average Heart Rate (bpm)'] = walking_data['Average Heart Rate (bpm)'].fillna(110)
running_data['Average Heart Rate (bpm)'] = running_data['Average Heart Rate (bpm)'].fillna(int(mean_hr_running))
cycling_data['Average Heart Rate (bpm)'] = cycling_data['Average Heart Rate (bpm)'].fillna(int(mean_hr_cycling))

# Count missing values in running data
display(running_data.isnull().sum())
```

```python
%matplotlib inline

# Import matplotlib, set style, and ignore specific warning
import matplotlib.pyplot as plt
import warnings

plt.style.use('ggplot')
warnings.filterwarnings(
    action='ignore', module='matplotlib.figure', category=UserWarning,
    message=('This figure includes Axes that are not compatible with tight_layout, so results might be incorrect.')
)

# 🔧 Sort index before slicing to avoid KeyError
running_data = running_data.sort_index()

# Prepare data: subset running activities from 2013 to 2018
runs_subset_2013_2018 = running_data.loc['2013':'2018']

# Create, plot and customize in one step
runs_subset_2013_2018.plot(y=['Distance (km)', 'Duration', 'Average Heart Rate (bpm)'],
                           subplots=True,
                           sharex=False,
                           figsize=(12, 16),
                           linestyle='none',
                           marker='o',
                           markersize=3,
                           title='Running Stats (2013–2018)'
                          )

# Show the plot
plt.show()
```

```python
# Prepare running data for the last 4 years (2015–2018)
runs_subset_2015_2018 = running_data.loc['2015':'2018']

# Calculate annual statistics (mean per year)
print('How my average run looks in last 4 years:')
annual_stats = runs_subset_2015_2018.resample('Y').mean(numeric_only=True)
display(annual_stats)

# Calculate weekly statistics (mean per week)
print('Weekly averages of last 4 years:')
weekly_stats = runs_subset_2015_2018.resample('W').mean(numeric_only=True)
display(weekly_stats)

# Mean weekly counts (number of training sessions per week)
weekly_counts_average = runs_subset_2015_2018.resample('W').size().mean()
print('How many trainings per week I had on average:', weekly_counts_average)
```

```python
# Fix: Ensure datetime index is correct
running_data.index = pd.to_datetime(running_data.index)
running_data = running_data.sort_index()

# Prepare data (note: forward years 2015 to 2018)
runs_subset_2015_2018 = running_data['2015':'2018']

# Extract distance and heart rate series
runs_distance = runs_subset_2015_2018['Distance (km)']
runs_hr = runs_subset_2015_2018['Average Heart Rate (bpm)']

# Create plot with two subplots
fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(12, 8), sharex=True)

# Plot and customize first subplot: Distance
runs_distance.plot(ax=ax1, color='orange', marker='o', linestyle='none', markersize=3)
ax1.set(ylabel='Distance (km)', title='Historical Data with Averages (2015–2018)')
ax1.axhline(runs_distance.mean(), color='blue', linewidth=1, linestyle='-.')  # mean line

# Plot and customize second subplot: Heart Rate
runs_hr.plot(ax=ax2, color='gray', marker='x', linestyle='none', markersize=3)
ax2.set(xlabel='Date', ylabel='Average Heart Rate (bpm)')
ax2.axhline(runs_hr.mean(), color='red', linewidth=1, linestyle='--')  # mean line

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Prepare data: total running distance per year
df_run_dist_annual = running_data['Distance (km)'].resample('Y').sum()

# Create plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot and customize
df_run_dist_annual.plot(ax=ax, marker='*', markersize=14, linewidth=0, color='blue')

ax.set(
    ylim=[0, 1210], 
    xlim=[pd.Timestamp('2012-12-31'), pd.Timestamp('2019-01-01')],
    ylabel='Distance (km)',
    xlabel='Years',
    title='Annual totals for distance'
)

# Highlight performance zones
ax.axhspan(1000, 1210, color='green', alpha=0.4, label='Excellent')
ax.axhspan(800, 1000, color='yellow', alpha=0.3, label='Good')
ax.axhspan(0, 800, color='red', alpha=0.1, label='Needs Improvement')

# Optional: add a legend
ax.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Import required libraries
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Prepare weekly data: sum of distance per week
df_run_dist_wkly = running_data['Distance (km)'].resample('W').sum()

# Decompose the time series into trend, seasonality, and residuals
decomposed = sm.tsa.seasonal_decompose(df_run_dist_wkly, model='additive', period=52, extrapolate_trend=1)

# Create plot
fig, ax = plt.subplots(figsize=(12, 6))

# Plot and customize
decomposed.trend.plot(ax=ax, label='Trend', linewidth=2)
decomposed.observed.plot(ax=ax, label='Observed', linewidth=0.5)

ax.legend()
ax.set_title('Running Distance Trend (Weekly)')

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Prepare heart rate data for all runs
hr_zones = [100, 125, 133, 142, 151, 173]
zone_names = ['Easy', 'Moderate', 'Hard', 'Very hard', 'Maximal']
zone_colors = ['green', 'yellow', 'orange', 'tomato', 'red']

# Select only valid heart rate values
df_run_hr_all = running_data['Average Heart Rate (bpm)'].dropna()

# Create plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot histogram and customize bin colors
n, bins, patches = ax.hist(df_run_hr_all, bins=hr_zones, alpha=0.5, edgecolor='black')

# Apply colors to each bin
for i in range(len(patches)):
    patches[i].set_facecolor(zone_colors[i])

# Set axis labels and title
ax.set(
    title='Distribution of Heart Rate Zones During Runs',
    ylabel='Number of Runs',
    xlabel='Average Heart Rate (bpm)'
)
ax.xaxis.set(ticks=hr_zones)

# Optional: add zone labels as legend
from matplotlib.patches import Patch
legend_handles = [Patch(facecolor=zone_colors[i], label=zone_names[i]) for i in range(len(zone_names))]
ax.legend(handles=legend_handles, title='HR Zones')

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Concatenating three DataFrames
df_run_walk_cycle = pd.concat([running_data, walking_data, cycling_data])

# Define columns of interest
dist_climb_cols = ['Distance (km)', 'Climb (m)']
speed_col = ['Average Speed (km/h)']

# Calculating total distance and climb per activity type
df_totals = df_run_walk_cycle.groupby('Type')[dist_climb_cols].sum()

print('Totals for different training types:')
display(df_totals)

# Calculating summary statistics (mean, std, min, etc.)
df_summary = df_run_walk_cycle.groupby('Type')[dist_climb_cols + speed_col].describe()

# Combine totals into the summary table under a 'total' sub-column
for col in dist_climb_cols:
    df_summary[(col, 'total')] = df_totals[col]

# Reorder columns to move 'total' to the end
df_summary = df_summary.sort_index(axis=1, level=0)

print('Summary statistics for different training types:')
display(df_summary)
```
